```python
import numpy as np
    
x = np.arange(-np.pi, np.pi, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
x=[1,2,3,4]
y=[3,1,7,1]
plt.plot(x,y,marker='+',linestyle='dashed',color='g',linewidth=4)
m=[5,3,6,4]
plt.plot(x,m,marker='o',color='orange')
plt.xlabel("over")
plt.ylabel("Run")
plt.legend(["KKR",'Rcb'], loc="upper right")
plt.title("Today match")
plt.show()
```

```python
#axes method[left,bottom,height,width] fraction
ax = plt.axes([0.1,0.1,0.8,0.8])
x= np.arange(1,10,0.1)
s = np.sin(x)
c= np.cos(x)
ax.plot(x,s,'g+:')
ax.plot(x,c,"rH-")
ax.legend(["Rcb","CSK"],loc="upper right")
ax.set_title("IPL Match")
```

```python
x=[1,2,3,4]
y=[2,3,4,5]
fg, ax= plt.subplots(2,2)
ax[0][0].plot(x,y)
ax[0][1].scatter(x,y)
ax[1][0].bar(x,y)
ax[1][1].hist(x)
ax[0][0].set_title("My line graph")
ax[0][1].set_title("My scatter graph")
ax[1][0].set_title("My bar graph")
ax[1][1].set_title("My hist graph")
plt.show()
```

```python
x=np.array([1,2,3,4])
fg,ax= plt.subplots(2,2)
ax[0][0].plot(x)
ax[0][1].plot(x**2)
ax[1][0].plot(x**3)
ax[1][1].plot(x**(0.5)) #square root
```

```python
from matplotlib import gridspec
help(gridspec)
```

```python
ax =gridspec.GridSpec(nrows =2,ncols=2,width_ratios=[2,1],height_ratios=[1,2])
fg= plt.figure()
first_plot= fg.add_subplot(ax[0])
first_plot.plot(x)
second_plot= fg.add_subplot(ax[1])
second_plot.plot(x)
third_plot= fg.add_subplot(ax[2])
third_plot.plot(x)
forth_plot= fg.add_subplot(ax[3])
forth_plot.plot(x)
```
